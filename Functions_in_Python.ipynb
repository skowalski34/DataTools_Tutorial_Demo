## Description
The fundamental units of any Python program are functions. A function is a named segment of code with a name that carries out a specific action. Usually, this entails receiving some input, processing it, and then producing an output. Functions let you divide your code into distinct, reusable pieces. Functions can even be used outside of the original application in which they were defined. There are two types of functions called built-in functions and user-defined functions. Built-in functions are built into Python. Whereas, user-defined functions are defined by the users themselves.
